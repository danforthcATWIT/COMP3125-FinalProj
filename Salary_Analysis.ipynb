```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% ! important; }<style>"))
```

```python
salary_data = pd.read_csv("ds_salaries.csv",  encoding="ISO-8859-1")
salary_data.head()
```

```python
#Question 1:
```

```python
avg_salary = salary_data.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending = False)
print(avg_salary)
```

```python
# Group by country and calculate mean salary
average_salary_by_country = salary_data.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False)

filtered_data = average_salary_by_country.head(45)
# Create a bar plot
plt.figure(figsize=(12, 6))
filtered_data.plot(kind='bar', color='skyblue')
plt.title('Average Salary by Country')
plt.xlabel('Country')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
from scipy.stats import f_oneway


# Perform ANOVA
f_statistic, p_value = f_oneway(*[group['salary_in_usd'] for name, group in salary_data.groupby('employee_residence')])

# Check if the p-value is significant
if p_value < 0.05:
    print("ANOVA Result: Significant differences exist.")


else:
    print("ANOVA Result: No significant differences found.")
```

```python
#Question 2 
```

```python
# Scatter plot with regression line
sns.regplot(x='remote_ratio', y='salary_in_usd', data=salary_data)
plt.title('Correlation between Remote Ratio and Salary')
plt.xlabel('Remote Ratio')
plt.ylabel('Salary (USD)')
plt.show()

# Calculate correlation coefficient
correlation_coefficient = salary_data['remote_ratio'].corr(salary_data['salary_in_usd'])
print(f"Correlation Coefficient: {correlation_coefficient}")
```

```python
# Group by 'remote_ratio' and calculate mean salary for each group
average_salary_by_remote_ratio = salary_data.groupby('remote_ratio')['salary_in_usd'].mean().reset_index()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.bar(average_salary_by_remote_ratio['remote_ratio'], average_salary_by_remote_ratio['salary_in_usd'])
plt.title('Average Salary by Remote Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('Average Salary (USD)')
plt.show()
```

```python
# Group data by 'remote_ratio' and extract salary values for each group
grouped_data = [group['salary_in_usd'] for name, group in salary_data.groupby('remote_ratio')]

# Perform ANOVA
f_statistic, p_value = f_oneway(*grouped_data)

# Check if the p-value is significant
if p_value < 0.05:
    print("ANOVA Result: Significant differences exist.")
else:
    print("ANOVA Result: No significant differences found.")
```

```python
# Calculate average salary for each job title
average_salary_by_job_title = salary_data.groupby('job_title')['salary_in_usd'].mean().reset_index()

# Select the top 50 job titles by average salary
top_50_job_titles = average_salary_by_job_title.nlargest(50, 'salary_in_usd')

# Bar plot to display average salary for the top 50 job titles
plt.figure(figsize=(12, 6))
sns.barplot(x='job_title', y='salary_in_usd', data=top_50_job_titles, palette='viridis')
plt.title('Top 50 Job Titles by Average Salary')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Group data by 'job_title' and extract salary values for each group
grouped_data_job_title = [group['salary_in_usd'] for name, group in salary_data.groupby('job_title')]

# Perform ANOVA
f_statistic_job_title, p_value_job_title = f_oneway(*grouped_data_job_title)

# Check if the p-value is significant
if p_value_job_title < 0.05:
    print("ANOVA Result for Job Titles: Significant differences exist.")
else:
    print("ANOVA Result for Job Titles: No significant differences found.")
```
